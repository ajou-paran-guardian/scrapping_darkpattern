```python
# keyword 예시 : ~남음, ~개, ~보고 있~, ~까지, ~명이 담음, ~보는 
```

```python
#pandas import
import pandas as pd
import time
import os

#BeautifulSoup import
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

#Selenium import
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException  # 예외 처리 추가
from selenium.webdriver.chrome.options import Options # user-agent 변경
```

```python
# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_argument("Mozilla/5.0")  # 원하는 User-Agent 문자열로 변경
```

```python
# 가장 작은 div 레이아웃 가져오기
def find_div(elements):
    html = []
    for element in elements:
        try:
            # 현재 요소를 포함하는 가장 가까운 div 찾기
            closest_div = element.find_element(By.XPATH, './ancestor::div[1]')
            html_code = closest_div.get_attribute('outerHTML')
            
            # 결과 출력
            print("HTML Code:")
            print(html_code)
            
            #결과값 저장
            html.append(html_code)
            
        except NoSuchElementException:
            print("가장 가까운 div 요소를 찾을 수 없습니다.")
            
    return html
```

```python
data = pd.read_csv('false_test.csv')
driver = webdriver.Chrome()

# 결과 파일 존재 여부 확인
output_file = "result.csv"
file_exists = os.path.isfile(output_file)

for url in data['URL']:
    #1. url을 연다
    print("Dynamic URL open : "+ url)
    driver.get(url) #동적 페이지
    
    time.sleep(3)
    
    #2. element 불러오기(동적)
    keyword = "남음"
    dynamic_elements = driver.find_elements(By.XPATH, f"//*[contains(text(), '{keyword}')]")
    
    #3. div 레이아웃 가져오기(동적)
    dynamic_html = find_div(dynamic_elements)
    #df = pd.DataFrame({'HTML':dynamic_html})
    
    #4. 정적 페이지에서 해당 레이아웃 찾기
    print("Static URL open: "+ url)
    static_html = urlopen(url) #정적페이지
    bs_obj = BeautifulSoup(static_html, "html.parser")
    
    #5. 정적 페이지에서 동적 페이지와 동일한 div 찾기
    for dynamic_html_code in dynamic_html:
        
        #동적 요소에서 텍스트만 추출
        dynamic_soup = BeautifulSoup(dynamic_html_code, "html.parser")
        dynamic_text = dynamic_soup.get_text(strip=True)
        
        #동적 요소의 텍스트를 기반으로 동일한 내용 찾기
        static_element = bs_obj.find(string=lambda text: dynamic_text in text if text else False)
        
        if static_element:
            static_parent = static_element.find_parent() #텍스트가 속한 부모 태그를 추출
            comparison_result = "False info"
            print("---동일한 정보가 존재합니다.---")
            print("------------------------")
            print(f"dynamic : {dynamic_html_code} | static : {static_parent}")
        else:
            print(f"---해당 정보가 없습니다.--- | dynamic : {dynamic_html_code}")
            compoarison_result = "True info"
            
        print(f" result : {comparison_result}")
        print("------------------------")
    
```
